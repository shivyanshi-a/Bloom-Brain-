```python
import joblib
import pandas as pd
import random
import time
import matplotlib.pyplot as plt

model = joblib.load("plant_rf_model.pkl")

moisture_list = []
temperature_list = []
humidity_list = []

plt.ion()

for i in range(20):

    moisture = random.randint(100, 400)
    temperature = random.randint(15, 40)
    humidity = random.randint(20, 90)
    light = random.randint(200, 800)
    ph = round(random.uniform(4.5, 7.5), 2)
    air_quality = random.randint(300, 800)
    nutrient = random.randint(20, 80)

    new_data = pd.DataFrame([[moisture, temperature, humidity, light, ph, air_quality, nutrient]],
        columns=['moisture','temperature','humidity','light','ph','air_quality','nutrient'])

    prediction = model.predict(new_data)[0]

    print("Prediction:", prediction)

    moisture_list.append(moisture)
    temperature_list.append(temperature)
    humidity_list.append(humidity)

    plt.clf()

    plt.subplot(3,1,1)
    plt.plot(moisture_list)
    plt.title("Moisture")

    plt.subplot(3,1,2)
    plt.plot(temperature_list)
    plt.title("Temperature")

    plt.subplot(3,1,3)
    plt.plot(humidity_list)
    plt.title("Humidity")

    plt.pause(1)

plt.ioff()
plt.show()
```
