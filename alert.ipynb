```python
import pandas as pd
import joblib

model = joblib.load("plant_rf_model.pkl")

def get_value(name):
    while True:
        try:
            value = float(input(f"Enter {name}: "))
            return value
        except ValueError:
            print("⚠️ Please enter a numeric value")

moisture = get_value("moisture")
temperature = get_value("temperature")
humidity = get_value("humidity")
light = get_value("light")
ph = get_value("pH")
air_quality = get_value("air quality")
nutrient = get_value("nutrient level")

new_data = pd.DataFrame([[moisture, temperature, humidity, light, ph, air_quality, nutrient]],
                        columns=['moisture','temperature','humidity','light','ph','air_quality','nutrient'])

prediction = model.predict(new_data)[0]

print("\n🌿 Plant says:")

if prediction == "water_needed":
    print("💧 Help! I'm thirsty... please give me some water 🥺")
elif prediction == "stress":
    print("😟 I'm feeling stressed... please check my environment")
else:
    print("😊 Yay! Your plant is happy now!")
```
