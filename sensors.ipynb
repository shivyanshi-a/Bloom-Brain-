```python
import joblib
import pandas as pd

# 1. Model load karein
model = joblib.load("plant_rf_model.pkl")

# 2. Ek "Perfect Happy" data banayein
# Dhayan dein: Columns wahi rakhein jo training mein the
happy_data = pd.DataFrame([[70, 24, 55, 450, 6.5, 20, 90]], 
    columns=['moisture', 'temperature', 'humidity', 'light', 'ph', 'air_quality', 'nutrient'])

# 3. Prediction check karein
prediction = model.predict(happy_data)[0]

print("-" * 30)
print(f"Model Predicted Value: {prediction}")
print(f"Data Type: {type(prediction)}")
print("-" * 30)
```
