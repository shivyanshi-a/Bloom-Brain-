```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000  # each class

# healthy
healthy = pd.DataFrame({
    "moisture": np.random.randint(250, 400, n),
    "temperature": np.random.randint(20, 28, n),
    "humidity": np.random.randint(50, 80, n),
    "light": np.random.randint(400, 700, n),
    "ph": np.round(np.random.uniform(6.0, 7.5, n), 2),
    "air_quality": np.random.randint(600, 800, n),
    "nutrient": np.random.randint(50, 90, n),
    "label": "healthy"
})

# stress
stress = pd.DataFrame({
    "moisture": np.random.randint(150, 250, n),
    "temperature": np.random.randint(28, 35, n),
    "humidity": np.random.randint(30, 50, n),
    "light": np.random.randint(300, 500, n),
    "ph": np.round(np.random.uniform(4.5, 6.0, n), 2),
    "air_quality": np.random.randint(500, 650, n),
    "nutrient": np.random.randint(30, 60, n),
    "label": "stress"
})

# water_needed
water = pd.DataFrame({
    "moisture": np.random.randint(80, 150, n),
    "temperature": np.random.randint(22, 32, n),
    "humidity": np.random.randint(20, 45, n),
    "light": np.random.randint(350, 600, n),
    "ph": np.round(np.random.uniform(5.0, 7.0, n), 2),
    "air_quality": np.random.randint(550, 700, n),
    "nutrient": np.random.randint(20, 50, n),
    "label": "water_needed"
})

# combine
data = pd.concat([healthy, stress, water])

# shuffle
data = data.sample(frac=1).reset_index(drop=True)

# save
data.to_csv("plant_data_balanced_3000.csv", index=False)

print("Balanced dataset created")
print(data['label'].value_counts())
```
